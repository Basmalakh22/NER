```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset from a CSV file
df = pd.read_csv('ner_datasetreference.csv', encoding='ISO-8859-1')
df.head(20)
```

```python
import pandas as pd
import string
from nltk.corpus import stopwords

df['clean_text']=df['Word'].str.lower()

STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in text.split() if word not in STOPWORDS])
df['clean_text'] = df['clean_text'].apply(lambda x: remove_stopwords(x))

import re
def remove_spl_chars(text):
    text = re.sub('[^a-zA-Z0-9]', ' ', text)
    text = re.sub('\s+', ' ', text)
    return text
df['clean_text'] = df['clean_text'].apply(lambda x: remove_spl_chars(x))

df.head(20)
```

```python
df['Word'] = df['Word'].apply(lambda x: ''.join(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vect = TfidfVectorizer()
features = tfidf_vect.fit_transform(df['Word'])
```

```python
pd.DataFrame(features.toarray(),columns = tfidf_vect.vocabulary_.keys())
```

```python
#Train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, df['Word'], test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
import numpy as np
# predicted = 0,1
predicted = model.predict(X_test)

# accuracy
print(np.mean(predicted == y_test ))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test ,predicted))

#[true negative ,false positive ,false negative ,predicted yes ]

```
