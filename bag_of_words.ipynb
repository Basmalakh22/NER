```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ner_datasetreference.csv', encoding='ISO-8859-1')
df.head()
```

```python
#return a Pandas Series with counts of unique words in the 'Word' column 
df['Word'].value_counts()
```

```python
#create a new column 'Dateline' and set it to 1 for rows where the 'Word' column is equal to 'Dateline' and 0 otherwise
df['Dateline'] = df['Word'].apply(lambda x: 1 if x =='Dateline' else 0)
```

```python
#Train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['Word'], df['Dateline'], test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_cv, y_train)
```
